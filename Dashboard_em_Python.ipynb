# **Dicionário de Dados**
It is mixed data, that is, both numerical and categorical data are present, The dataset consists of 1000 students record with a total of eight features, viz...

1. gender:categorical
2. race/ethnicity: categorical
3. parental level of education: categorical
4. lunch: categorical
5. test preparation course: categorical
6. math score: numerical
7. reading score: numerical; and
8. writing score: numerical

The dataset is taken from 'Student Performance in Exames - Marks secured by the student in various subjects' by Jakki Seshapanpu Ref. -

https://www.kaggle.com/spscientist/students-performance-in-exams

```python
import sweetviz as sv
import pandas as pd
```

# Carregando Dataset

```python
#Loading the Data
data=pd.read_csv('/content/StudentsPerformance.csv')

#Printing data insights
print(data.info)
```

```python
# Printing few data samples
data.head(5)
```

```python
# Analyzing data
report=sv.analyze(data)

# Generating report
report.show_notebook('eda_report.html')
report.show_notebook(w=900, h="Full", scale=0.8)

```

```python
# Split the data into datasets
data1=data[0:400]
data2=data[400:]
```

```python
print(data1.info())
```

```python
print(data2.info())
```

```python
report_comp=sv.compare(data1,data2)
report_comp.show_html()
```
